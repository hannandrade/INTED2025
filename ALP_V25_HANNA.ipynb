4 - Crie um algoritmo que receba uma idade do usuário, e depois verifique em qual faixa etária está a idade:

até 12 - Você é uma criança!

de 13 a 18 - Você é um Adolescente!

de 19 a 60 - Você é um adulto!

acima de 60 - Você é idoso!

```python
idade = int(input("Digite sua idade: "))

if idade <= 12:
  print("Você é uma criança!")
elif idade >= 13 and idade <= 18:
  print("Você é um Adolescente!")
elif idade >= 19 and idade <= 60:
  print("Você é um adulto!")
else:
  print("Você é idoso!")
```
